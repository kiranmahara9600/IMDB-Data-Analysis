#**Let's check which actor has acted in the most number of movies. Deep diving into the movies, genres and profits corresponding to this actor.**

```python
#-----------------------------------Block-1--------------------------------------

#defining a function to evaluate arbitrary Python expressions from a string-based input 
def convert_to_list(str):
  return eval(str)   #eval() is used to evaluate arbitrary Python expressions from a string-based input 
```

```python
#-----------------------------------Block-2--------------------------------------

#applying the function defined in Block-1 in non null observations of 'cast' feature 
imdb_df['cast']=imdb_df[~imdb_df['cast'].isna()]['cast'].apply(convert_to_list)  
```

* **To find actor who acted in most number of movies**

```python
#-----------------------------------Block-3--------------------------------------

# finding the actors list and then the actor who did maximum no. of movies
actors_list=[]   #initialising an empty list to store actor names
for i in range(3000):
  cast_list=[]   
  if type(imdb_df.loc[i,'cast'])==list:   #putting a condition because there are some na values which is of floating type that we dont want 
    cast_list=cast_list+imdb_df.loc[i,'cast']   #storing cast observations in cast_list 
    for elem in cast_list:   #iterating over cast_list to find actor name
      actors_list.append(elem['name'])   #storing name of actor in actors_list 
      
actors_df=pd.Series(actors_list)   #converting actors_list into a Series
#checking for no. of times an actor name has appeared in actors_df
actors_numofmovies_df=actors_df.value_counts().sort_values(ascending=False).reset_index()
actors_numofmovies_df.rename(columns={'index':'actor name',0:'num of movies'},inplace=True)   #renaming column name
actors_numofmovies_df[0:5]   #top 5 actors having maximum no. of movies
```

Robert De Niro and Samuel L. Jackson acted in most number of movies.

* **Movies corresponding to these two actors**

```python
#-----------------------------------Block-4--------------------------------------

#finding the movie names in which 'Robert De Niro' and 'Samuel L. Jackson' acted
#initialising empty lists
cast_list=[]   #to store cast observations
robert_movie_list=[]   #to store 'Robert De Niro' movies
samuel_movie_list=[]   #to store 'Samuel L. Jackson' movies

for i in range(3000):
  if type(imdb_df.loc[i,'cast'])==list:    #putting a condition because there are some na values which is of floating type that we dont want   
    cast_list=imdb_df.loc[i,'cast']   #storing cast observations in cast_list
    for elem in cast_list:    #iterating over cast_list to find actor name
      if 'Robert De Niro' in elem['name']:   #putting condition that if actor name is 'Robert De Niro' 
        robert_movie_list.append(imdb_df.loc[i,'original_title'])   #storing movies of 'Robert De Niro' in robert_movie_list
      if 'Samuel L. Jackson' in elem['name']:    #putting condition that if actor name is 'Samuel L. Jackson' 
        samuel_movie_list.append(imdb_df.loc[i,'original_title'])   #storing movies of 'Samuel L. Jackson' in samuel_movie_list
        
#converting lists into pandas series       
robert_movie_data=pd.Series(robert_movie_list)   
samuel_movie_data=pd.Series(samuel_movie_list)
actor_movies_df=pd.concat([robert_movie_data,samuel_movie_data],axis=1)   #merging series to get movies corresponding to these two actors
actor_movies_df.rename(columns={0:'Robert De Niro-movies',1:'Samuel L. Jackson-movies'},inplace=True)   #renaming columns
actor_movies_df
```

```python
#-----------------------------------Block-5--------------------------------------

#applying the function defined in Block-1 in non null observations of 'genres' feature
imdb_df['genres']=imdb_df[~imdb_df['genres'].isna()]['genres'].apply(convert_to_list)
```

* **Genres of movies corresponding to these two actors**

```python
#-----------------------------------Block-6--------------------------------------

#finding genres associated with the actors who did most number of movies
#initialising empty lists
cast_list=[]   #to store cast observations
robert_genres_list=[]   #to store genres of 'Robert De Niro' movies
samuel_genres_list=[]   #to store genres of 'Samuel L. Jackson' movies

for i in range(3000):
  if type(imdb_df.loc[i,'cast'])==list:   #putting a condition because there are some na values which is of floating type that we dont want   
    cast_list=imdb_df.loc[i,'cast']
    if type(imdb_df.loc[i,'genres'])==list:
      for elem in cast_list:     #iterating over cast_list to find actor name
        if 'Robert De Niro' in elem['name']:    #putting condition that if actor name is 'Robert De Niro' 
          for dictionary in imdb_df.loc[i,'genres']: #iterating over imdb_df.loc[i,'genres'] to find genres associated with 'Robert De Niro' movies
            robert_genres_list.append(dictionary['name'])   #storing movie genres of 'Robert De Niro' in robert_genres_list
        if 'Samuel L. Jackson' in elem['name']:   #putting condition that if actor name is 'Samuel L. Jackson' 
          for dictionary in imdb_df.loc[i,'genres']:#iterating over imdb_df.loc[i,'genres'] to find genres associated with 'Samuel L. Jackson' movies
            samuel_genres_list.append(dictionary['name'])    #storing movie genres of 'Samuel L. Jackson' in samuel_genres_list    
                   
robert_genres=pd.Series(robert_genres_list)   #converting lists into pandas series 
Robert_genres=robert_genres.value_counts().reset_index()   #count of different genres associated with Robert De Niro movies
Robert_genres.rename(columns={'index':'Robert De Niro-movies genres',0:'count of movies'},inplace=True)
samuel_genres=pd.Series(samuel_genres_list)
Samuel_genres=samuel_genres.value_counts().reset_index()   #count of different genres associated with Samuel L. Jackson movies
Samuel_genres.rename(columns={'index':'Samuel L. Jackson-movies genres',0:'count of movies'},inplace=True)
```

```python
#-----------------------------------Block-7--------------------------------------

#defining a function to calculate profit for given budget and revenue
def get_profit(bug,rev):
  pro=rev-bug
  return pro

#applying the function in imdb_df to create a new column 'profit'
imdb_df['profit']=imdb_df.apply(lambda x: get_profit(x['budget'],x['revenue']),axis=1)  
```

* **Profit corresponding to these two actor movies**

```python
#-----------------------------------Block-8--------------------------------------

#finding profit corresponding to these two actors movies
#initialising empty lists
cast_list=[]   #to store cast observations
robert_profit_list=[]   #to store profit corresponsding to 'Robert De Niro' movies 
samuel_profit_list=[]   #to store profit corresponsding to 'Samuel L. Jackson' movies 

for i in range(3000):
  if type(imdb_df.loc[i,'cast'])==list:    #putting a condition because there are some na values which is of floating type that we dont want
    cast_list=imdb_df.loc[i,'cast']
    for elem in cast_list:   #iterating over cast_list to find actor name
      if 'Robert De Niro' in elem['name']:    #putting condition that if actor name is 'Robert De Niro' 
        robert_profit_list.append(imdb_df.loc[i,'profit'])   #storing profit across 'Robert De Niro' in robert_profit_list
      if 'Samuel L. Jackson' in elem['name']:    #putting condition that if actor name is 'Samuel L. Jackson' 
        samuel_profit_list.append(imdb_df.loc[i,'profit'])   #storing profit across 'Samuel L. Jackson' in samuel_profit_list

robert_profit_df=pd.Series(robert_profit_list)   #converting to pandas series
samuel_profit_df=pd.Series(samuel_profit_list)
#merging series to get movies and corresponding profit for these two actors
actors_movieprofit_df=pd.concat([robert_movie_data,robert_profit_df,samuel_movie_data,samuel_profit_df],axis=1)  
actors_movieprofit_df.rename(columns={0:'Robert De Niro-movies',1:'Robert De Niro-movies profit',2:'Samuel L. Jackson-movies',
                                      3:'Samuel L. Jackson-movies profit'},inplace=True)     #renaming column names                                     
actors_movieprofit_df
```

**Outcomes**

* *Robert De Niro and Samuel L. Jackson are the top 2 actors who have acted in most number of movies.*

* *Drama, Crime, Comedy, Thriller are the most preferred movie genres by Robert De Niro.*

* *Action, Crime, Thriller,Adventure, science Fiction are the most preferred movie genres by Samuel L. Jackson.*

* *These both actors have losses corresponding to some of movies.*

#**Let's check for top 3 directors and actors preferred by these directors**

* **Top 3 directors**

```python
#-----------------------------------Block-9--------------------------------------

#finding the directos list and then top 3 directors
#applying function defined in Block-1 in non null observations of 'crew' feature
imdb_df['crew']=imdb_df[~imdb_df['crew'].isnull()]['crew'].apply(convert_to_list)
director_list=[]   #intitialising an empty list to store director names

for i in range(3000):
  if type(imdb_df.loc[i,'crew'])==list:   #because there are some na values which is of floating type that we dont want
    for elem in imdb_df.loc[i,'crew']:   #iterating over imdb_df.loc[i,'crew'] to get director information
      if elem['job']=='Director':   
        director_list.append(elem['name'])   #storing director name in director_list  

director_df=pd.Series(director_list)   #conversion to pandas series
director_numofmovies_df=director_df.value_counts().sort_values(ascending=False).reset_index()   #count of movies corresponding to different directors
director_numofmovies_df.rename(columns={'index':'director name',0:'num of movies'},inplace=True)
director_numofmovies_df[0:3]   #top 3 directors                             
```

Clint Eastwood, Ron Howard and Steven Spielberg	 are the top three directors.

* **Actors preferred by top 3 directors**

```python
#-----------------------------------Block-10--------------------------------------

#initialising empty lists
director_list=[]   #to store directors name
Clint_actors_list=[]   #to store actor names preferred by 'Clint Eastwood' 
Ron_actors_list=[]    #to store actor names preferred by 'Ron Howard' 
Steven_actors_list=[]    #to store actor names preferred by 'Steven Spielberg'

for i in range(3000):
  if (type(imdb_df.loc[i,'crew'])==list) & (type(imdb_df.loc[i,'cast'])==list): #because there are some na values that we dont want
    for elem in imdb_df.loc[i,'crew']:    #iterating over imdb_df.loc[i,'crew'] to get director information
      if elem['job']=='Director':
        if 'Clint Eastwood' in  elem['name']:   #putting a condition to get actor choosen by this actor for his movies
          for elem in imdb_df.loc[i,'cast']:
            Clint_actors_list.append(elem['name'])   #storing actors name (preferred by 'Clint Eastwood' ) in Clint_actors_list 
        if 'Ron Howard'in elem['name']:   #putting a condition to get actor choosen by this actor for his movies
          for elem in imdb_df.loc[i,'cast']:
            Ron_actors_list.append(elem['name'])   #storing actor name (preferred by 'Ron Howard') in Ron_actors_list
        if 'Steven Spielberg'in elem['name']:   #putting a condition to get actor choosen by this actor for his movies
          for elem in imdb_df.loc[i,'cast']:
            Steven_actors_list.append(elem['name'])    #storing actor name (preferred by 'Steven Spielberg') in Steven_actors_list  

#converting lists to series and then counting no. of times an actor is choosen 
Clint_actors_df=pd.Series(Clint_actors_list).value_counts().sort_values(ascending=False).reset_index()  
Clint_actors_df.rename(columns={'index':'actors',0:'Top 3 actors preferred by Clint Eastwood'},inplace=True) 

Ron_actors_df=pd.Series(Ron_actors_list).value_counts().sort_values(ascending=False).reset_index()
Ron_actors_df.rename(columns={'index':'actors',0:'Top 3 actors preferred by Ron Howard'},inplace=True) 

Steven_actors_df=pd.Series(Steven_actors_list).value_counts().sort_values(ascending=False).reset_index()
Steven_actors_df.rename(columns={'index':'actors',0:'Top 3 actors preferred by Steven Spielberg'},inplace=True) 
```
